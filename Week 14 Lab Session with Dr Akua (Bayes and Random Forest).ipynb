## Naive Bayes 

```python
#importing libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB

df=pd.read_csv(r"C:\Users\LENOVO\Downloads\golfdataset.csv")
```

```python
df
```

```python
#encoding the data; string to numeric
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
df['Outlook']=le.fit_transform(df.Outlook.values)
df['Temp']=le.fit_transform(df.Temp.values)
df['Humidity']=le.fit_transform(df.Humidity.values)
df['Windy']=le.fit_transform(df.Windy.values)
df['PlayGolf']=le.fit_transform(df.PlayGolf.values)

df
```

```python
#spliting 
label=df['PlayGolf']
features=df.drop('PlayGolf', axis=1)

x_train, x_test, y_train, y_test =train_test_split(features, label, train_size=0.80, random_state=0)
```

```python
#traing the model on the training set
gnb=GaussianNB()
gnb.fit(x_train, y_train)
```

```python
# making prediction on the testing set
#y_pred=gnb.predict([[2,0,1,0]])
#y_pred
```

```python
# making prediction on the testing set
#y_pred=gnb.predict([[0,0,0,0]])
y_pred=gnb.predict(x_test)
y_pred
```

```python
#camparing
from sklearn import metrics
print('Gaussian Naive Bayes model accuracy is',metrics.accuracy_score(y_test, y_pred)*100,'%')
```
